# Telco Customer Churn Analysis

### Telco is a telecommunications company that provides phone and internet services to its customers. Telco is interested in retaining their customer base while reducing the amount of customer churn. Churn refers to the loss of customers. Inspecting aspects and services provided by Telco and their relationship to churn could provide some useful business insights.

```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df = pd.read_csv(r"C:\Users\singh\OneDrive\Documents\Customer_Churn.csv")
```

### Preprocessing and Cleaning Data

```python
df.dtypes
```

#### These are the various columns in the data set and the variables we will explore to see if there is any significant trends in churn. 

```python
for column in df:
    print(f'{column} : {df[column].unique()}')
```

#### We can see that Total Charges is an object type instead of a float. Let's change it to a integer to avoid any future errors. We can also see that the Senior Citizen column is currently mapped to '0 and 1' instead of 'Yes' and 'No'.  Let's change it to avoid any confusion.

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce', downcast='float')
df["SeniorCitizen"]= df["SeniorCitizen"].map({0: "No", 1: "Yes"})
```

#### Let us begin our analysis by first identifying how much churn Telco is experiencing and visualize it below.

```python
churn_pct = 100* df['Churn'].value_counts()/len(df['Churn'])
labels=['No','Yes']

plt.pie(churn_pct, labels=labels,autopct='%.0f%%')
```

#### We see that there is about 27% of the customer base that is churning. Let us examine all the different areas of our data base and focus in on which variables are the most important.

### Churn by Demographics

#### Let's take a look at the demographic-related variables in the data set and how they may relate to the amount of churn Telco experiences. We'll specifically examine any trends in churn related to gender, partner status, dependent status, and age.

```python

plt.figure(figsize = (15, 15))
sns.set(style='whitegrid')
palette = {
    'No': 'tab:blue',
    'Yes': 'tab:orange'}
palette2= {'Male': 'tab:blue',
    'Female': 'tab:orange'}

plt.subplot(3, 2, 1)
ax = sns.countplot(x="Churn", hue="gender", palette=palette2,data=df);
ax.set_ylabel('Number of Customers', fontsize = 10)
ax.set_xlabel('Churn', fontsize = 10)
ax.set_xticklabels(['Not Churned','Churned'])
ax.set_title('Gender Churn', fontsize=12)

plt.subplot(3, 2, 2)
ax = sns.countplot(x="Churn", hue="SeniorCitizen",palette=palette,data=df);
ax.set_ylabel('Number of Customers', fontsize = 10)
ax.set_xlabel('Churn', fontsize = 10)
ax.set_xticklabels(['Not Churned','Churned'])
ax.set_title('Senior Citizen Churn', fontsize=12)

plt.subplot(3, 2, 3)
ax = sns.countplot(x="Churn", hue="Partner",palette=palette, data=df);
ax.set_ylabel('Number of Customers', fontsize = 10)
ax.set_xlabel('Churn', fontsize = 10)
ax.set_xticklabels(['Not Churned','Churned'])
ax.set_title('Partner Churn', fontsize=12)


plt.subplot(3, 2, 4)
ax = sns.countplot(x="Churn", hue="Dependents",palette=palette,data=df);
ax.set_ylabel('Number of Customers', fontsize = 10)
ax.set_xlabel('Churn', fontsize = 10)
ax.set_xticklabels(['Not Churned','Churned'])
ax.set_title('Dependents Churn', fontsize=12)
plt.show()
```

#### In terms of demographics, Gender is not very indicative of churn.  Customers that don't have partners are more likely to churn. Customers without dependents are also more likely to churn.

#### Telco should offer discounts to household packages and contracts to target more families as opposed to singles. Telco has a fair amount of churn with its senior citizen demographic, which may mean more senior discounts or contracts with features that are not needed. 

### Churn through Technological Services

#### Next let's look at the technological services provided by Telco to determine if there is churn happening in that sector. Let us start with Phone Services before moving onto Internet Services.

```python
plt.figure(figsize = (15, 5))
sns.set(style='whitegrid')
palette = {
    'No': 'tab:blue',
    'Yes': 'tab:orange',
    'No phone service':'tab:green'}


plt.subplot(1, 2, 1)
ax = sns.countplot(x="Churn", hue="PhoneService", palette=palette,data=df);
ax.set_ylabel('Number of Customers', fontsize = 10)
ax.set_xticklabels(['Not Churned','Churned'])
ax.set_title('Phone Service Churn', fontsize=12)

plt.subplot(1, 2, 2)
ax = sns.countplot(x="Churn", hue="MultipleLines", palette=palette,data=df);
ax.set_ylabel('Number of Customers', fontsize = 10)
ax.set_xticklabels(['Not Churned','Churned'])
ax.set_title('Multiple Lines Churn', fontsize=12)
plt.show()
```

#### From these graphs we can see that phone service is very common among all customers. Customers with multiple lines have a slightly higher churn rate but overall are not a prominent factor for churn.  Phone service must remain a default feature as part of Telco contracts.

#### Now let's look at the different internet services provided by Telco and see how churn relates to these particular factors. Let us start with the type of Internet Service causes the most amount of churn (DSL, Fiber Optic, No internet). 

```python

plt.figure(figsize = (15, 5))
sns.set(style='whitegrid')

palette = {
    'DSL': 'tab:blue',
    'Fiber optic': 'tab:orange',
    'No':'tab:green'}

ax = sns.countplot(x="Churn", hue="InternetService", palette=palette,data=df);
ax.set_ylabel('Number of Customers', fontsize = 10)
ax.set_xticklabels(['Not Churned','Churned'])
ax.set_title('Churn By Internet Service', fontsize=12)
plt.show()
```

#### Customers with fiber optic internet are more probable to churn than those with DSL connection. Those with DSL or no internet service altogether have much lower churn rates. Fiber Optic is a more sophisticated type of internet service and therefore demands higher costs. 

#### Telco should consider lowering the price point for Fiber Optic internet in the future to reduce churn. Now let's move onto additional tech features.

```python
palette = {
    'No': 'tab:blue',
    'Yes': 'tab:orange',
    'No internet service':'green'}

plt.figure(figsize = (15, 5))
sns.set(style='whitegrid')
plt.subplot(1, 3, 1)
ax = sns.countplot(x="Churn", hue="OnlineSecurity", palette=palette,data=df);
ax.set_ylabel('Number of Customers', fontsize = 10)
ax.set_xlabel('Churn', fontsize = 10)
ax.set_xticklabels(['Not Churned','Churned'])
ax.set_title('Online Security Churn', fontsize=12)

plt.subplot(1, 3, 2)
ax = sns.countplot(x="Churn", hue="OnlineBackup", palette=palette, data=df);
ax.set_ylabel('Number of Customers', fontsize = 10)
ax.set_xlabel('Churn', fontsize = 10)
ax.set_xticklabels(['Not Churned','Churned'])
ax.set_title('Online Backup Churn', fontsize=12)

plt.subplot(1, 3, 3)
ax = sns.countplot(x="Churn", hue="DeviceProtection", palette=palette,data=df);
ax.set_ylabel('Number of Customers', fontsize = 10)
ax.set_xlabel('Churn', fontsize = 10)
ax.set_xticklabels(['Not Churned','Churned'])
ax.set_title('Device Protection Churn', fontsize=12)
plt.show()
```

#### Customers without these three features (Online Security, Online Backup, Device Protection) are more likely to churn than with customers who do have those features. Telco should make sure these three are included in most of their packages to ensure customer retention. Let's next look at streaming services.

```python
palette = {
    'Yes': 'tab:orange',
    'No': 'tab:blue',
    'No internet service':'green'}

plt.figure(figsize = (15, 5))
plt.subplot(1, 2, 1)
ax = sns.countplot(x="Churn", hue="StreamingTV", palette=palette,data=df);
ax.set_ylabel('Number of Customers', fontsize = 10)
ax.set_xticklabels(['Not Churned','Churned'])
ax.set_title('TV Streaming Churn', fontsize=12)

plt.subplot(1, 2, 2)
ax = sns.countplot(x="Churn", hue="StreamingMovies", palette=palette,data=df);
ax.set_ylabel('Number of Customers', fontsize = 10)
ax.set_xticklabels(['Not Churned','Churned'])
ax.set_title('Movie Streaming Churn ', fontsize=12)
plt.show()
```

#### As we can see above, both streaming services are not particularly predictive for churn. Let's finish the services analysis by examining customer services like tech support availibility and paperless billing. 

```python
palette = {
    'Yes': 'tab:blue',
    'No': 'tab:orange',
    'No internet service':'green'}

plt.figure(figsize = (15, 5))
sns.set(style='whitegrid')
plt.subplot(1, 2, 1)
ax = sns.countplot(x="Churn", hue="PaperlessBilling", palette=palette, data=df);
ax.set_ylabel('Number of Customers', fontsize = 10)
ax.set_xticklabels(['Not Churned','Churned'])
ax.set_title('Churn By Paperless Billing', fontsize=12)

plt.subplot(1, 2, 2)
ax = sns.countplot(x="Churn", hue="TechSupport", palette=palette,data=df);
ax.set_ylabel('Number of Customers', fontsize = 10)
ax.set_xticklabels(['Not Churned','Churned'])
ax.set_title('Churn By Tech Support', fontsize=12)
plt.show()
```

Telco should revisit paperless billing as an optional feature to see if it is worth including for most packages it is currently offering. Telco should definitely ensure it has tech support availibilty to its customers to help retention.

### Churn through Contract Type 

#### Let us looks at the three different Contract Types (Month-to-Month, One Year, Two Year) that are available and see if there are any noticeable trends. 

```python
palette = {
    'Month-to-month': 'tab:blue',
    'One year': 'tab:orange',
    'Two year':'green'}

plt.figure(figsize = (15, 5))
sns.set(style='whitegrid')
ax = sns.countplot(x="Churn", hue="Contract", palette=palette,data=df);
ax.set_ylabel('Number of Customers', fontsize = 10)
ax.set_xticklabels(['Not Churned','Churned'])
ax.set_title('Contract Type and Churn', fontsize=14)
plt.show()
```

#### The customers who have a month to month contract have a very high churn rate, especially compared to the one-year or two-year contracts. Telco should offer additional popular features or a discounted price for monthly contracts to increase retention.

### Customer Behavior

#### Let's examine how customer's preferred choice of payment may contribute to churn as well as length of tenure.

```python
palette = {
    'Electronic check': 'tab:blue',
    'Mailed check': 'tab:orange',
    'Bank transfer (automatic)':'tab:green',
    'Credit card (automatic)':'tab:red'}

plt.figure(figsize = (15, 5))
sns.set(style='whitegrid')
ax = sns.countplot(x="Churn", hue="PaymentMethod", palette=palette,data=df);
ax.set_ylabel('Number of Customers', fontsize = 10)
ax.set_xticklabels(['Not Churned','Churned'])
ax.set_title('Churn By Payment Method', fontsize=14)
plt.show()
```

#### Electronic Check has the highest churn rate of all payment methods. Telco should look into builing a better infrastructure to help facilitate transactions with E-checks.  Let's see if customer tenure has any relationship to churn.

```python
plt.figure(figsize=(15,5))
sns.set(style='whitegrid')
ax = sns.histplot(data=df, x='tenure', hue='Churn', binwidth=2, kde=True)
ax.set_xlabel('Tenure (number of months)', fontsize = 10)
ax.set_ylabel('Number of customers', fontsize = 10)
plt.title('Customer Tenure')
plt.show()
```

#### From the graphic we can see that churned customers are mostly those who had been a customer for only 010 months. We can also see very high retention in the 70 month range. This means Telco should focus on retaining those customers early as there is a greater chance that they remain customers in the long run.

### Monthly and Total Charges

#### Now we'll see how both monthly and total charges impact churn.

```python
sns.set_context("paper",font_scale=1.1)
plt.figure(figsize=(15,5))
sns.set(style='whitegrid')
ax = sns.kdeplot(df.MonthlyCharges[(df["Churn"] == 'No') ],
                color="Blue", shade = True);
ax = sns.kdeplot(df.MonthlyCharges[(df["Churn"] == 'Yes') ],
                ax =ax, color="Orange", shade= True);
ax.legend(["Not Churned","Churned"],loc='upper right');
ax.set_ylabel('Density');
ax.set_xlabel('Monthly Charges');
ax.set_title('Monthly Charges and Churn');
```

#### As far as monthly charges is related to churn, most customers who churn are the ones who are charged between 80 and 100. Conversely the most retained customer base are the ones who are charged between 0-40 monthly. Upfront monthly costs should be kept lowered to attract new customers and to retain as well.  Let's look at Total Charges next.

```python
sns.set_context("paper",font_scale=1.1)
plt.figure(figsize=(15,5))
sns.set(style='whitegrid')
ax = sns.kdeplot(df.TotalCharges[(df["Churn"] == 'No') ],
                color="Blue", shade = True);
ax = sns.kdeplot(df.TotalCharges[(df["Churn"] == 'Yes') ],
                ax =ax, color="Orange", shade= True);
ax.legend(["Not Churned","Churned"],loc='upper right');
ax.set_ylabel('Density');
ax.set_xlabel('Total Charges');
ax.set_title('Total Charges and Churn');
```

#### Hightest Churn takes place within the 0-2000 range, which suggests that the highest churn is occuring from the early customers, which is corroborated with the charts above.

### Overall correlation and Key Takeaways

#### Let's map every variable against churn to see any correlation between the features we have been looking at.

```python
df_2 = df.iloc[:,1:]
df_2['Churn'].replace(to_replace='Yes', value=1, inplace=True)
df_2['Churn'].replace(to_replace='No',  value=0, inplace=True)
df_dummies = pd.get_dummies(df_2)
```

```python
plt.figure(figsize=(15,8))
df_dummies.corr()['Churn'].sort_values(ascending = False).plot(kind='bar')
```

#### Features with positive correlation to churn indicate features that are driving churn. Features with negative correlation to churn indicate features that are driving retention.


#### We can see that month-to-month contracts is very positively correlated with churn, which reinforces the idea of retaining an early customer base. The absence of key features such as online security, device protection, tech support, and online backup seem to be most positively correlated with churn which is consistent with our previous analysis. Fiber Optic has a high correlation with churn as does Electronic Check Payments.

#### Tenure and two-year contracts are most negatively correlated with churn as are the aforementioned key features like online security, device protection, tech support, and online backup. 

#### Final Recommendations: include the key tech features above, reduce month-to-month costs to attract new customers. Reduce costs for fiber optic internet, and create a better infrastructure to facilitate electronic check payments. These will all increase retention and therefore increase tenure, which is the key to builing a customer base with little churn.
